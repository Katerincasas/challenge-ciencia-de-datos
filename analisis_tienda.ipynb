```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Sumatoria del valor total facturado por tienda
facturacion_t1 = tienda["Precio"].sum()
facturacion_t2 = tienda2["Precio"].sum()
facturacion_t3 = tienda3["Precio"].sum()
facturacion_t4 = tienda4["Precio"].sum()
```

```python
print("Facturación total:")
print(f"Tienda 1: S/. {facturacion_t1:.2f}")
print(f"Tienda 2: S/. {facturacion_t2:.2f}")
print(f"Tienda 3: S/. {facturacion_t3:.2f}")
print(f"Tienda 4: S/. {facturacion_t4:.2f}")
```

# 2. Ventas por categoría

```python
# Conteo de ventas por categoría en cada tienda
print("Ventas por categoría - Tienda 1")
print(tienda['Categoría del Producto'].value_counts())

print("\nVentas por categoría - Tienda 2")
print(tienda2['Categoría del Producto'].value_counts())

print("\nVentas por categoría - Tienda 3")
print(tienda2['Categoría del Producto'].value_counts())

print("\nVentas por categoría - Tienda 4")
print(tienda2['Categoría del Producto'].value_counts())
```

```python
facturacion = [facturacion_t1, facturacion_t2, facturacion_t3, facturacion_t4]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

plt.figure(figsize=(10, 6))
bars = plt.bar(tiendas, facturacion, color=['#4CAF50', '#2196F3', '#FF9800', '#E91E63'])
plt.xlabel('Tiendas', fontsize=12)
plt.ylabel('Ingresos (S/.)', fontsize=12)
plt.title('Ingresos Totales por Tienda', fontsize=14)
plt.grid(axis='y', linestyle='--')

# Añadir etiquetas con formato de moneda
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height,
             f'S/. {height:,.2f}',
             ha='center', va='bottom')

plt.show()
```

```python
# prompt: grafico de  Calificaciones de Clientes de las 4 tiendas

import matplotlib.pyplot as plt

# Calificaciones promedio de cada tienda
calificaciones_promedio = {
    'Tienda 1': tienda['Calificación'].mean(),
    'Tienda 2': tienda2['Calificación'].mean(),
    'Tienda 3': tienda3['Calificación'].mean(),
    'Tienda 4': tienda4['Calificación'].mean()
}

tiendas = list(calificaciones_promedio.keys())
calificaciones = list(calificaciones_promedio.values())

plt.figure(figsize=(10, 6))
plt.bar(tiendas, calificaciones, color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
plt.xlabel("Tiendas")
plt.ylabel("Calificación Promedio")
plt.title("Calificaciones Promedio de Clientes por Tienda")
plt.ylim(0, 5)  # Ajustar el límite del eje y para una mejor visualización

# Agregar etiquetas de valor encima de cada barra
for i, v in enumerate(calificaciones):
    plt.text(i, v + 0.1, f"{v:.2f}", ha='center', va='bottom', fontsize=10)

plt.show()
```

```python
# prompt: genera grafico de  Productos Más Vendidos en las 4 tiendas , coloca una etiqueta con el valor en cada barra

import pandas as pd
import matplotlib.pyplot as plt

# ... (previous code)

# Productos más vendidos por tienda (top 5)
def plot_top_products(df, tienda_num):
    top_products = df['Producto'].value_counts().head(5)
    plt.figure(figsize=(10, 6))
    bars = plt.bar(top_products.index, top_products.values, color='skyblue')
    plt.xlabel("Productos")
    plt.ylabel("Cantidad Vendida")
    plt.title(f"Productos Más Vendidos - Tienda {tienda_num}")
    plt.xticks(rotation=45, ha='right')  # Rotar etiquetas del eje x
    plt.tight_layout() # Ajustar el diseño

    # Agregar etiquetas de valor encima de cada barra
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, yval, ha='center', va='bottom')

    plt.show()


plot_top_products(tienda, 1)
plot_top_products(tienda2, 2)
plot_top_products(tienda3, 3)
plot_top_products(tienda4, 4)
```
