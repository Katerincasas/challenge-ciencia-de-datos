```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Sumatoria del valor total facturado por tienda
facturacion_t1 = tienda["Precio"].sum()
facturacion_t2 = tienda2["Precio"].sum()
facturacion_t3 = tienda3["Precio"].sum()
facturacion_t4 = tienda4["Precio"].sum()
```

```python
print("Facturación total:")
print(f"Tienda 1: S/. {facturacion_t1:.2f}")
print(f"Tienda 2: S/. {facturacion_t2:.2f}")
print(f"Tienda 3: S/. {facturacion_t3:.2f}")
print(f"Tienda 4: S/. {facturacion_t4:.2f}")
```

```python
# prompt: grafico de  Calificaciones de Clientes de las 4 tiendas

import matplotlib.pyplot as plt

# Calificaciones promedio de cada tienda
calificaciones_promedio = {
    'Tienda 1': tienda['Calificación'].mean(),
    'Tienda 2': tienda2['Calificación'].mean(),
    'Tienda 3': tienda3['Calificación'].mean(),
    'Tienda 4': tienda4['Calificación'].mean()
}

tiendas = list(calificaciones_promedio.keys())
calificaciones = list(calificaciones_promedio.values())

plt.figure(figsize=(10, 6))
plt.bar(tiendas, calificaciones, color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
plt.xlabel("Tiendas")
plt.ylabel("Calificación Promedio")
plt.title("Calificaciones Promedio de Clientes por Tienda")
plt.ylim(0, 5)  # Ajustar el límite del eje y para una mejor visualización

# Agregar etiquetas de valor encima de cada barra
for i, v in enumerate(calificaciones):
    plt.text(i, v + 0.1, f"{v:.2f}", ha='center', va='bottom', fontsize=10)

plt.show()
```

```python
# prompt:  grafico de Productos Más Vendidos y menos vendidos por tiendas  responsive, coloca el valor como etiqueta en cada barra

import matplotlib.pyplot as plt

def plot_top_bottom_products(df, tienda_name):
    top_products = df['Producto'].value_counts().head(5)
    bottom_products = df['Producto'].value_counts().tail(5)

    fig, axes = plt.subplots(1, 2, figsize=(14, 6))
    fig.suptitle(f'Productos Más y Menos Vendidos - {tienda_name}', fontsize=16)


    axes[0].bar(top_products.index, top_products.values, color='skyblue')
    axes[0].set_title('Top 5 Productos')
    axes[0].set_xlabel('Producto')
    axes[0].set_ylabel('Cantidad Vendida')
    axes[0].tick_params(axis='x', rotation=45)
    for i, v in enumerate(top_products.values):
        axes[0].text(i, v + 1, str(v), ha='center', va='bottom')


    axes[1].bar(bottom_products.index, bottom_products.values, color='lightcoral')
    axes[1].set_title('Bottom 5 Productos')
    axes[1].set_xlabel('Producto')
    axes[1].set_ylabel('Cantidad Vendida')
    axes[1].tick_params(axis='x', rotation=45)
    for i, v in enumerate(bottom_products.values):
        axes[1].text(i, v + 0.1, str(v), ha='center', va='bottom')

    plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Ajusta el layout para evitar superposiciones
    plt.show()

plot_top_bottom_products(tienda, 'Tienda 1')
plot_top_bottom_products(tienda2, 'Tienda 2')
plot_top_bottom_products(tienda3, 'Tienda 3')
plot_top_bottom_products(tienda4, 'Tienda 4')
```
